```python
#Limpieza datos#

#Se utiliza en la biblioteca Pandas para reemplazar valores faltantes (representados como NaN) en un DataFrame#
#Sirve para limpiar datos y preparar un conjunto de datos para el análisis, ya sea reemplazando los valores nulos con un valor específico (como 0)#

df.fillna(0, inplace=True)
df
```

```python
#Se genera un nuevo DataFrame, con la limpieza de los NaN hecha anteriormente# 

df2 = df.copy()
df2
```

```python
#Para datos atípicos se necesita el boxplot numérico, se convertira la columna en este caso a numéricas#

df2['Engine size'] = pd.to_numeric(df['Engine size'], errors='coerce').fillna(0)
df2['Year of manufacture'] = pd.to_numeric(df['Year of manufacture'], errors='coerce').fillna(0)
df2["Price"]=pd.to_numeric(df['Price'], errors='coerce').fillna(0)
df2["Mileage"]=pd.to_numeric(df['Mileage'], errors='coerce').fillna(0)
df2
```

```python
df2.info()
```

```python
#Convertir las palabras a minicusculas, esto me sirve porque evita duplicados y mejora la consistencia al agrupar o contar categorías#

for col in ['Manufacturer', 'Model', 'Fuel type']:
    df2[col] = df2[col].str.strip().str.lower()
df2
```

```python
#Funciona para identificar si hay palabras extrañas o raras dentro de estas columnas en especifico#

print(df2['Manufacturer'].unique())
print(df2['Fuel type'].unique())
print(df2['Model'].unique())
```

```python
#Busca palabras raras o valores extraños# 

import numpy as np
valores_invalidos = ['?', 'unknown', 'none', 'null', '-', '_', 'n/a', 'na', 'nan', '', ' ','ayuder$']
df2.replace(valores_invalidos, np.nan, inplace=True)
df2.fillna('desconocido', inplace=True) 
df2
```

```python
#Se revisa que ya no se encuntren las palabaras extrañas en los datos de texto# 

print(df2['Manufacturer'].unique())
print(df2['Fuel type'].unique())
print(df2['Model'].unique())
```

```python
#Elimina caracteres raros o símbolos#
#Quita espacios sobrantes#
#Convierte todo a minúsculas# 

import re

def limpiar_texto(texto):
    if isinstance(texto, str):
        texto = re.sub(r'[^A-Za-z0-9\s\-]', '', texto)  # deja letras, números y guiones
        texto = re.sub(r'\s+', ' ', texto).strip()
        return texto.lower()
    return texto

for col in df2.select_dtypes(include='object').columns:
    df2[col] = df2[col].apply(limpiar_texto)
df2
```
